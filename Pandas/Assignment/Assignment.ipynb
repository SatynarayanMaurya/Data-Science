```python
import pandas as pd
```

```python
df = pd.read_table("https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv")
```

```python
df
```

```python
# Remove the dollar sign and convert to float
df['item_price'] = df['item_price'].str.replace('$','').astype(float)
df['item_price'].head()
```

```python
# We are going to add more column 
df['total_price'] = df['quantity'].astype(float) * df['item_price'].astype(float)
df.head()
```

```python
# A small data cleaning we are going to remove those rows whose choice description is NaN
df.dtypes
```

```python
df = df.dropna(subset=['choice_description'])
```

```python
df.head()
```

```python
df['choice_description'].head()
```

```python
# Go with another file 
df1 = pd.read_csv("https://raw.githubusercontent.com/vincentarelbundock/Rdatasets/master/datasets.csv")
df1.head(10)
```

```python
# Here let's we have to change the column name Cols to columns 
df1.rename(columns={"Cols":"Columns"},inplace=True)
```

```python
df1.head(10)
```

```python
# Lets we want to convert the title into upper case
df1['Title'] = df1['Title'].str.upper()
df1.head()
```

```python
df1['Rows'] = df1['Rows'].fillna(0)  # This will remove the NaN values to 0
```

```python
df1.head()
```

```python
# We are going to sort according to the Columns
sorted_df = df1.sort_values(by="Columns", ascending=True)
sorted_df.head()
```

```python
# Now Move to the next File 
df2 = pd.read_table("https://raw.githubusercontent.com/justmarkham/DAT8/master/data/beer.txt",sep=" ")
df2.head()
```

```python
df2['name'] = df2['name'].str.lower()
df2.head()
```

```python
df2['name'] = df2['name'].str.capitalize()
df2.head()
```

```python
df2['cost'] = df2['cost'].round(3)
df2.head()
```

```python
# Create New column alcohol per calories
df2['alcohol_per_calories'] = df2['alcohol'] / df2['calories']
df2.head()
```

```python
# We can sort the table according to the column sodium
sorted_by_sodium = df2.sort_values(by="sodium")
sorted_by_sodium.head()
```
