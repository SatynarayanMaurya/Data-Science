```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame(np.random.randn(10,4), index =pd.date_range('2025-11-06',periods=10), columns=['a','b','c','d'])
df
```

```python
df['new'] = df['b'] + df['d']
df
```

```python
df['sqrt_of_d'] = df['d'].apply(lambda x : x**2)
df
```

```python
def find_log (x):
    if x < 0 :
        return x
    else:
        return np.log(x)
```

```python
df['log_of_d'] = df['d'].apply(find_log)
```

```python
df
```

```python
def test1 (x):
    if x < 0:
        return 0
    else:
        return 1
```

```python
df['new_a'] = df['a'].apply(test1)
df
```
