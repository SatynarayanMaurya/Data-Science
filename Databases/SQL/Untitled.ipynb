```python
import mysql.connector as connector 
try:
    myDb = connector.connect(host="localhost",database="satyasql",user="root", passwd="S@ty@9838021627", use_pure=True)
    cur = myDb.cursor()
    for i in range(10):
        query = "INSERT INTO students VALUES (%s, %s, %s, %s)"
        values = (i + 2, 'Satynarayan Maurya'+str(i), 21150 + i + 1, 'B.Tech')
        cur.execute(query, values)
    myDb.commit()
    myDb.close()
    print("Data inserted successfully")
except Exception as e:
    myDb.close()
    print(str(e))
```

```python
# Fetching data from DB
import mysql.connector as connector
try:
    myDb = connector.connect(host="localhost", user="root", database="satyasql",passwd="S@ty@9838021627",use_pure=True)
    cur = myDb.cursor()
    cur.execute("select * from students")
    for i in cur.fetchall():
        print(i)
    print("All data are fetched successfully")
    myDb.close()
except Exception as e:
    myDb.close()
    print(str(e))
```
