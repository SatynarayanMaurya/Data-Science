```python
class Bank:
    def __init__ (self,name,amount):
        self.name = name;
        self.amount = amount;

    def depositeAmount (self,amount):
        self.amount += amount;
        return print(f"Your total amount is : {self.amount}")

    def withdrawAmount (self,amount):
        if(self.amount < amount):
            return print(f"You have not enough Money for withdraw. You total amount is : {self.amount}")
        else:
            self.amount -= amount;
            return print(f"Your remaining amount is : {self.amount}")

    def knowYourBalance (self):
        return print(f"Your total amount is : {self.amount}")
```

```python
satya = Bank("Satya",1000)
```

```python
satya.knowYourBalance()
```

```python
satya.withdrawAmount(4000)
```
