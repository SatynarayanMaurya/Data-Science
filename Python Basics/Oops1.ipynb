```python
class data:
    def __init__(self,fileName,fileType,data,size):
        self.fileName = fileName
        self.fileType = fileType
        self.data = data
        self.size = size

    def openFile (self):
        try:
            f = open(self.fileName,'w+')
        except Exception as e:
            print("Can not found the file : ",e)
        else:
            f.write("This is the file content")
        finally :
            f.close()
            
    def readFile (self):
        try:
            f = open(self.fileName,'r')
        except Exception as e:
            print("Can not read the file : ",e)
        else :
            content = f.read();
            print(content)
        finally:
            f.close()
```

```python
newFile = data("text.txt","Text File","",0)
```
